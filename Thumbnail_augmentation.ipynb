```python
# adding thumbnail link to the dataset
import requests
import pandas as pd
import time
from urllib.parse import urljoin
import json
import os
#from dotenv import load_dotenv
```

```python
movies=pd.read_csv('final_data.csv')
```

```python
class MovieThumbnailFinder:
    def __init__(self, tmdb_api_key=None, omdb_api_key=None):
        self.tmdb_api_key = tmdb_api_key
        self.omdb_api_key = omdb_api_key
        self.tmdb_base_url = "https://api.themoviedb.org/3"
        self.tmdb_image_base_url = "https://image.tmdb.org/t/p/w500"
        self.omdb_base_url = "http://www.omdbapi.com/"

    def search_tmdb(self, title, year=None):
        """Search for movie using TMDb API"""
        if not self.tmdb_api_key:
            return None

        params = {
            'api_key': self.tmdb_api_key,
            'query': title
        }

        if year:
            params['year'] = year

        try:
            response = requests.get(f"{self.tmdb_base_url}/search/movie", params=params)
            response.raise_for_status()
            data = response.json()

            if data['results']:
                poster_path = data['results'][0].get('poster_path')
                if poster_path:
                    return f"{self.tmdb_image_base_url}{poster_path}"
            return None

        except requests.RequestException as e:
            print(f"TMDb error for {title}: {e}")
            return None

    def search_omdb(self, title, year=None):
        """Search for movie using OMDb API"""
        if not self.omdb_api_key:
            return None

        params = {
            'apikey': self.omdb_api_key,
            't': title
        }

        if year:
            params['y'] = year

        try:
            response = requests.get(self.omdb_base_url, params=params)
            response.raise_for_status()
            data = response.json()

            if data.get('Response') == 'True':
                poster_url = data.get('Poster')
                if poster_url and poster_url != 'N/A':
                    return poster_url
            return None

        except requests.RequestException as e:
            print(f"OMDb error for {title}: {e}")
            return None

    def find_thumbnail(self, title, year=None):
        """Find thumbnail for a movie, trying TMDb first, then OMDb"""
        # Try TMDb first
        if self.tmdb_api_key:
            thumbnail_url = self.search_tmdb(title, year)
            if thumbnail_url:
                return thumbnail_url

        # Try OMDb as fallback
        if self.omdb_api_key:
            return self.search_omdb(title, year)

        return None

    def process_dataset(self, movies_df, title_column='title', year_column=None,
                       output_column='thumbnail_url', delay=0.25):
        """Process a dataset of movies to find thumbnails"""
        results = []
        successful = 0

        for idx, row in movies_df.iterrows():
            title = row[title_column]
            year = row[year_column] if year_column and year_column in row else None

            print(f"Processing {idx + 1}/{len(movies_df)}: {title}")

            thumbnail_url = self.find_thumbnail(title, year)
            results.append(thumbnail_url)

            if thumbnail_url:
                successful += 1
                print(f"  ✓ Found thumbnail")
            else:
                print(f"  ✗ No thumbnail found")

            time.sleep(delay)

        print(f"\nFound thumbnails for {successful}/{len(movies_df)} movies")

        movies_df[output_column] = results
        return movies_df
```
