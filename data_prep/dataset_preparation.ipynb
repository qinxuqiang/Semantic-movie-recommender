```python
import pandas as pd
import numpy as np
import json
```

```python
movies_original=pd.read_csv('tmdb_5000_movies.csv')
credits_original=pd.read_csv('tmdb_5000_credits.csv')
movies=movies_original.copy()
credits=credits_original.copy()
```

```python
movies.head()
```

```python
# Columns 'genres','keywords' and 'production-companies' come in json string type. We extract the relevant info
def extract_name(json_string):
  data=json.loads(json_string)
  name_list=[entry['name'] for entry in data]
  return name_list

for col in ['genres','keywords','production_companies']:
  movies[col+'_cleaned']=movies[col].apply(extract_name)

```

```python
# Extract the country code from the column 'production_countries', language code from the column 'spoken_language'

def extract_language(json_string):
  data=json.loads(json_string)
  iso_list=[entry['iso_639_1'] for entry in data]
  return iso_list

movies['spoken_languages_cleaned']=movies['spoken_languages'].apply(extract_language)

def extract_country(json_string):
  data=json.loads(json_string)
  iso_list=[entry['iso_3166_1'] for entry in data]
  return iso_list

movies['production_countries_cleaned']=movies['production_countries'].apply(extract_country)
```

```python
# convert runtime in minutes to hour-min string for display
movies['runtime'].fillna(0,inplace=True)
def min_to_hour(n:float):
  n=int(n)
  hours= n//60
  minutes=n%60
  return f'{hours}h {minutes}min'
movies['runtime_str']=movies['runtime'].apply(min_to_hour)
```

```python
#movies.drop(['genres','keywords','production_companies','spoken_languages','production_countries'],axis=1,inplace=True)
```

```python
movies.drop(['homepage','original_title','status'],axis=1,inplace=True)
```

```python
# find row where release_date is nan and remove it
movies.drop(movies[movies['release_date'].isna()].index,axis=0,inplace=True)
```

```python
movies['year']=movies['release_date'].str[:4].astype('int')
```

```python
movies.isnull().sum()
```

```python
movies['budget in millions']=movies['budget']//1000000
movies['revenue in millions']=movies['revenue']//1000000
movies.drop(['budget','revenue'],axis=1,inplace=True)
```

```python
# json string extraction for credits
def extract_director(json_string):
  data=json.loads(json_string)
  director_list=[entry['name'] for entry in data if entry['job']=='Director']
  return director_list

credits['directors']=credits['crew'].apply(extract_director)

def extract_cast(json_string):
  data=json.loads(json_string)
  cast_list=[entry['name'] for entry in data]
  if len(cast_list)>5:
    cast_list=cast_list[:5]
  return cast_list

credits['cast']=credits['cast'].apply(extract_cast)

credits.drop(['crew','movie_id'],axis=1,inplace=True)
```

```python
#merge movie and credit based on 'title'
df=pd.merge(movies,credits,on='title')
```

```python
df["tagged_description"]=df[['id','title','directors','cast','genres_cleaned','keywords_cleaned','overview','tagline']].astype(str).agg(" ".join,axis=1)
```

```python
df.head()
```

```python
json_columns=['genres_cleaned','keywords_cleaned','production_countries_cleaned','spoken_languages_cleaned','cast','directors']
for col in json_columns:
  df[col]=df[col].apply(json.dumps)
```

```python
df.to_csv('final_data.csv',index=False)
```
